```python
book = soup.find_all('li', class_='col-xs-6 col-sm-4 col-md-3 col-lg-3')
```

```python
name = []
cost = []
stock = []
ratings = []
for i in book:
    name.append(i.find('h3').text.strip())
    cost.append(i.find('p', class_='price_color').text.strip())
    stock.append(i.find('p', class_='instock availability').text.strip())
    ratings.append(str(i.find('p', class_='star-rating')).split()[2][0:-2])

d = {
    'Name': name,
    'Cost': cost,
    'Stock Availability': stock,
    'Ratings': ratings
}

df = pd.DataFrame(d)
df
```

```python
final = pd.DataFrame()
for j in range(1, 51):

    url = f'http://books.toscrape.com/catalogue/page-{i}.html'
    webpage=requests.get(url, headers=headers).text
    
    headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
    webpage=requests.get('http://books.toscrape.com/catalogue/page-1.html', headers=headers).text
    soup = BeautifulSoup(webpage, 'lxml')
    book = soup.find_all('li', class_='col-xs-6 col-sm-4 col-md-3 col-lg-3')

    name = []
    cost = []
    stock = []
    ratings = []
    for i in book:
        name.append(i.find('h3').text.strip())
        cost.append(i.find('p', class_='price_color').text.strip())
        stock.append(i.find('p', class_='instock availability').text.strip())
        ratings.append(str(i.find('p', class_='star-rating')).split()[2][0:-2])
    
    d = {
        'Name': name,
        'Cost': cost,
        'Stock Availability': stock,
        'Ratings': ratings
    }
    
    df = pd.DataFrame(d)

    final = pd.concat([final, df], ignore_index=True)
    

    
```
