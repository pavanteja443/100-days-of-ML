```python
import requests
import pandas as pd
import time

df = pd.DataFrame()

def fetch_page(i, retries=10):
    for attempt in range(retries):
        try:
            response = requests.get(
                f'https://api.themoviedb.org/3/movie/top_rated?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page={i}'
            )
            if response.status_code == 200:
                return response.json()
            else:
                print(f"Attempt {attempt+1}: Failed with {response.status_code}")
        except Exception as e:
            print(f"Attempt {attempt+1}: {e}")
        time.sleep(2)
    return None

for i in range(1, 429):
    print(f"Fetching page {i}")
    result = fetch_page(i)
    if result:
        temp_df = pd.DataFrame(result['results'])[
            ['id', 'title', 'overview', 'release_date', 'popularity', 'vote_average', 'vote_count']
        ]
        df = pd.concat([df, temp_df], ignore_index=True)
    else:
        print(f"Page {i} failed after 10 retries.")
    time.sleep(1)
```
