## 2. Bar Plot (Numerical - Categorical)

```python
titanic.head()
```

## Pclass and Age relationship: Categorical-Numerical

```python
# generally x-axis: cateogorical, y-axos: numerical
sns.barplot(x = 'Pclass', hue = 'Pclass', y = 'Age', data = titanic, palette = 'Set1')
```

`Inferences: `  
`Average age of Pclass1: about 38`  
`Average age of Pclass2: about 30`  
`Average afe of Pclass3: about 25`

## Pclass and Fare relationship: Categorical-Numerical

```python
sns.barplot(x = 'Pclass', y = 'Fare', data = titanic, hue = 'Pclass', palette = 'Set1')
```

`Inferences: `  
`There is very less difference between average fare of Pclass2 and Pclass3`

```python
sns.barplot(x = 'Pclass', y = 'Fare', data = titanic, hue = 'Sex', palette = 'Set1')
```

```python
sns.barplot(x = 'Pclass', y = 'Age', data = titanic, hue = 'Sex', palette = 'Set1')
```

## 3. Box Plot (Numerical - Categorical)

```python
sns.boxplot(x = 'Sex', y = 'Age', data = titanic, hue= 'Survived')
```

## 4. Distplot (Numerical - Categorical)

```python
sns.distplot(titanic['Age'])
```

```python
sns.distplot(titanic[titanic['Survived']==0]['Age'], kde=True, hist=False)
sns.distplot(titanic[titanic['Survived']==1]['Age'], kde=True, hist=False)
```

```python
sns.histplot(titanic['Age'], kde = True)
```

## 5. HeatMap (Categorical - Categorical)

```python
titanic.head(3)
```

```python
# Pclass vs Survived
pd.crosstab(titanic['Pclass'], titanic['Survived'])
```

```python
# Pclass vs Survived
sns.heatmap(pd.crosstab(titanic['Pclass'], titanic['Survived']))
```

```python
titanic.groupby('Pclass').mean(numeric_only=True)['Survived']*100
```

```python
(titanic.groupby('Pclass').mean(numeric_only=True)['Survived']*100).plot(kind='bar')
```

```python
print(titanic.groupby('Sex').mean(numeric_only=True)['Survived']*100)
(titanic.groupby('Sex').mean(numeric_only=True)['Survived']*100).plot(kind='bar')
```

```python
print(titanic.groupby('Embarked').mean(numeric_only=True)['Survived']*100)
(titanic.groupby('Embarked').mean(numeric_only=True)['Survived']*100).plot(kind='bar')
```

## 6. ClusterMap (Categorical - Categorical)
